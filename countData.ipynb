```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
import sqlite3 as sq
```

```python
conn = sq.connect('registration.sqlite')
conn1 = sq.connect('station.sqlite')
conn2 = sq.connect('count.sqlite')
```

### CSV Imports

```python
#import the GDP csv
csv_file = "data/reg_data.csv"
registrationDf = pd.read_csv(csv_file, encoding ="ISO-8859-1")
registrationDf.head()
```

```python
registrationDf.columns
```

```python
registrationDfCopy= registrationDf[['State', ' Aprx_Registration_Count ']].copy()
```

```python
registrationDfCopy.to_sql(name='registration', con=conn, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from registration', con=conn)
```

```python
#import the countries csv
csv_file = "data/station_data.csv"
stationDf = pd.read_csv(csv_file, encoding ="ISO-8859-1")
stationDf.head()
```

```python
#preview the count
stationDf.columns
```

```python
Station_Count = stationDf.groupby(["State"]).count()["Station_Name"]
```

```python
#import the GDP csv
csv_file = "data/Count.csv"
CountDf = pd.read_csv(csv_file, encoding ="ISO-8859-1")
CountDf.head()
```

```python
CountDf.columns
```

```python
CountDf.to_sql(name='count', con=conn2, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from count', con=conn2)
```
